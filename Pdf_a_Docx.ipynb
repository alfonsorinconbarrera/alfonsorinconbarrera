```python
# Por estudioIAmbiental.com

# Instalar la biblioteca necesaria
!pip install pdf2docx

from pdf2docx import Converter
from google.colab import files
import os
import fitz  # PyMuPDF, se instala automáticamente con pdf2docx

# Subir el archivo PDF
print("Por favor, sube tu archivo PDF cuando aparezca el botón...")
uploaded = files.upload()

# Obtener el nombre del archivo subido
pdf_filename = list(uploaded.keys())[0]

def convertir_pdf_a_word(ruta_pdf, start=0, end=None):
    """
    Convierte un archivo PDF a formato Word (DOCX) en Google Colab con opciones mejoradas

    Args:
        ruta_pdf (str): Nombre del archivo PDF a convertir
        start (int): Página inicial para la conversión (0-based)
        end (int): Página final para la conversión (None = hasta el final)
    """
    try:
        # Crear nombre para el archivo Word
        ruta_word = os.path.splitext(ruta_pdf)[0] + '.docx'

        # Obtener número de páginas usando PyMuPDF
        doc = fitz.open(ruta_pdf)
        num_pages = len(doc)
        doc.close()

        if end is None:
            end = num_pages

        print(f"El documento tiene {num_pages} páginas. Convirtiendo desde página {start+1} hasta {end}...")

        # Crear el objeto convertidor y realizar la conversión
        cv = Converter(ruta_pdf)
        cv.convert(ruta_word, start=start, end=end,
                  kwargs={
                      'parse_tables': True,       # Mejorar detección de tablas
                      'keep_layout': True,        # Mantener diseño original
                      'debug': False,             # Desactivar debug
                      'multi_processing': True,   # Usar procesamiento múltiple
                      'min_text_len': 5,          # Texto mínimo para procesar
                  })

        # Cerrar el convertidor
        cv.close()

        print(f"✅ Conversión exitosa. Descargando archivo: {ruta_word}")

        # Descargar el archivo convertido
        files.download(ruta_word)

    except Exception as e:
        print(f"❌ Error durante la conversión: {str(e)}")

# Ejecutar la conversión
if pdf_filename.lower().endswith('.pdf'):
    print(f"Procesando {pdf_filename}...")
    convertir_pdf_a_word(pdf_filename)
else:
    print("Por favor, sube un archivo PDF válido.")
```
